```python
# Funções de Desenho de mapa

def labirinto():
    
    labirinto = [["     ", "XXXXX", "     ", "     ", "     ", "     ", "     "], 
                 ["     ", "     ", "XXXXX", "     ", "     ", "     ", "     "],
                ["     ", "XXXXX", "XXXXX", "     ", "     ", "     ", "     "],
                ["     ", "     ", "     ", "     ", "     ", "     ", "     "],
                ["     ", "     ", "     ", "     ", "     ", "     ", "Saida"],
                 ["     ", "     ", "     ", "     ", "     ", "     ", "     "],
                ["     ", "     ", "     ", "     ", "     ", "XXXXX", "     "]
                ]

    return labirinto

def desenha_labirinto(robo):
    
    lab = copy.deepcopy(labirinto())
    
    x, y = robo

    if verifica_chegada(robo) is True:
        lab[x][y] += " do robo"
        for linha in lab:
            print(linha)
        sys.exit()  # Interrompe a execução do código
        
    else:
        lab[x][y] =  "robo "
    
    for linha in lab:
        print(linha)
```

```python
## Função de verificar se o proximo passo é possivel

def verifica(robo):
    
    lab = copy.deepcopy(labirinto()) 
    x, y = robo

    
    if (x <=-1 or x >= len(lab) or y <= -1 or y >= len(lab)):
        return False

    if lab[x][y] == "XXXXX":
        return False
    
    else:
        return True
    
def verifica_chegada(robo):
    
    lab = copy.deepcopy(labirinto()) 
    x, y = robo
    
    if lab[x][y] == "Saida":
        return True
    
    else:
        return False
    
```

```python
# Funções de Movimento Basica

def direita(robo):
    
    x, y = robo
    
    robo = x , y+1
    
    return robo

def esquerda(robo):
    
    x, y = robo
    
    probo = x , y-1
    
    return probo

def emcima(robo):
    
    x, y = robo
    
    probo = x-1 , y
    
    return probo

def embaixo(robo):
    
    x, y = robo
    
    probo = x+1 , y
    
    return probo
```

```python
## Funções de Estrategia

def main_v1(robo):
    
    if verifica(direita(robo)):          #pd ir pra direita
        return direita(robo), 1
        
    elif verifica(embaixo(robo)): #pd ir pra baixo
        return embaixo(robo), 2
    
    elif verifica(emcima(robo)): #pd ir pra cima
        return emcima(robo), 3
    
    elif verifica(esquerda(robo)): #pd ir pra esquerda
         return esquerda(robo), 4
    
    else:
          return [-1, -1],5
          
def main_v2(robo):
          
    if verifica(direita(robo)): #pd ir pra direita
        return direita(robo), 1
            
    elif verifica(emcima(robo)): #pd ir pra cima
        return emcima(robo), 3  
          
    elif verifica(embaixo(robo)): #pd ir pra baixo
        return embaixo(robo),2
    
    elif verifica(esquerda(robo)): #pd ir pra esquerda
         return esquerda(robo), 4
    
    else:
          return [-1, -1], 5
        
def mudanca_estrategia(robo, i, movimentos):
    
    x,y = robo
    lab = copy.deepcopy(labirinto()) 

    if x > math.floor(len(lab)/2):     # Se tiver na area superior, vai para baixo
        
        while(True):
            if verifica(emcima(robo)) is False:
                return robo,i , 3
            i+=1
            robo = emcima(robo)
            x,y = robo
            chego = verifica_chegada(robo)
            movimentos.append(0)
            print("Iteração {} para emcima com a coordenada na mudança para v2 {}{}\n".format(i, x, y))

            desenha_labirinto(robo)
            
                    
    else:                              # Se tiver na area inferior, vai para cima
        while(True):

            if verifica(embaixo(robo)) is False:
                return robo, i , 2
            i+=1
            robo = embaixo(robo)
            x,y = robo
            chego = verifica_chegada(robo)
            movimentos.append(0)
            
            print("Iteração {} para embaixo com a coordenada na mudança para v1 {}{}\n".format(i, x, y))

            desenha_labirinto(robo)
            
```
